```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/Delhi_AQIBulletins.csv')

# Clean column names
df.columns = df.columns.str.strip().str.lower()

df.head()
```

```python
# Convert date
df['date'] = pd.to_datetime(df['date'], errors='coerce')

# Sort data
df = df.sort_values('date').reset_index(drop=True)

# Handle missing values
df['index value'] = df['index value'].ffill()

# Drop invalid rows
df = df.dropna(subset=['date'])

df.head()
```

```python
df['month'] = df['date'].dt.month
df['year'] = df['date'].dt.year

df.head()
```

```python
monthly_avg = df.groupby('month')['index value'].mean().sort_index()

print("Monthly Average AQI:\n")
print(monthly_avg)
```

```python
plt.figure(figsize=(10,5))

plt.plot(monthly_avg.index, monthly_avg.values, marker='o')

plt.title("Monthly Average AQI Trend")
plt.xlabel("Month")
plt.ylabel("Average AQI")

plt.grid()
plt.show()
```

```python
def categorize_aqi(aqi):
    if aqi < 100:
        return "Clean"
    elif aqi < 200:
        return "Moderate"
    else:
        return "High Pollution"

monthly_category = monthly_avg.apply(categorize_aqi)

print("\nMonthly Pollution Category:\n")
print(monthly_category)
```

```python
high_pollution_months = monthly_avg[monthly_avg > 200]

print("High Pollution Months:\n")
print(high_pollution_months)
```
