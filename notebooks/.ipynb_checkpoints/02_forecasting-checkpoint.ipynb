```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv('../data/Delhi_AQIBulletins.csv')

# Clean column names
df.columns = df.columns.str.strip().str.lower()

df.head()
```

```python
df['date'] = pd.to_datetime(df['date'], errors='coerce')
df = df.sort_values('date').reset_index(drop=True)

# Handle missing values
df['index value'] = df['index value'].ffill()

# Drop invalid rows
df = df.dropna(subset=['date'])

df.head()
```

```python
# Rolling features
df['rolling_mean_3'] = df['index value'].rolling(3).mean()
df['rolling_mean_7'] = df['index value'].rolling(7).mean()

# Drop NaN created by rolling
df = df.dropna()

df.head()
```

```python
features = df[['index value', 'rolling_mean_3', 'rolling_mean_7']].values
```

```python
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(features)
```

```python
def create_dataset(data, window=10):
    X, y = [], []
    for i in range(len(data) - window):
        X.append(data[i:i+window].flatten())
        y.append(data[i+window][0])  # target = index value
    return np.array(X), np.array(y)

X, y = create_dataset(scaled_data, window=10)

print("X shape:", X.shape)
print("y shape:", y.shape)
```

```python
split = int(0.8 * len(X))

X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]
```

```python
model = RandomForestRegressor(
    n_estimators=200,
    max_depth=15,
    min_samples_split=5,
    random_state=42
)

model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
# Only inverse transform index value (first column)
y_test_actual = scaler.inverse_transform(
    np.concatenate([y_test.reshape(-1,1), np.zeros((len(y_test),2))], axis=1)
)[:,0]

y_pred_actual = scaler.inverse_transform(
    np.concatenate([y_pred.reshape(-1,1), np.zeros((len(y_pred),2))], axis=1)
)[:,0]
```

```python
mae = mean_absolute_error(y_test_actual, y_pred_actual)
rmse = np.sqrt(mean_squared_error(y_test_actual, y_pred_actual))

print("Improved Model Performance:")
print("MAE:", round(mae, 2))
print("RMSE:", round(rmse, 2))
```

```python
plt.figure(figsize=(12,5))

plt.plot(y_test_actual, label='Actual AQI')
plt.plot(y_pred_actual, label='Predicted AQI')

plt.title("Improved AQI Forecasting")
plt.xlabel("Time")
plt.ylabel("AQI")

plt.legend()
plt.grid()

plt.show()
```

```python
last_window = scaled_data[-10:]
last_window = last_window.reshape(1, -1)

pred_scaled = model.predict(last_window)

pred_actual = scaler.inverse_transform(
    np.concatenate([pred_scaled.reshape(-1,1), np.zeros((1,2))], axis=1)
)[:,0]

print("Predicted AQI for next day:", round(pred_actual[0], 2))
```

### Model Improvement Summary

- Added rolling mean features (3-day & 7-day)
- Increased window size to 10 days
- Applied normalization for better stability
- Tuned Random Forest parameters

### Results

- MAE improved significantly
- RMSE reduced compared to baseline

### Insight

The model captures AQI trends more effectively after feature engineering.
