```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('../data/Delhi_AQIBulletins.csv')

# Standardize column names
df.columns = df.columns.str.strip().str.lower()

df.head()
```

```python
# Convert date column
df['date'] = pd.to_datetime(df['date'], errors='coerce')

# Sort by date
df = df.sort_values('date').reset_index(drop=True)

# Handle missing values
df['index value'] = df['index value'].ffill()

# Drop rows where date is missing
df = df.dropna(subset=['date'])

df.head()
```

```python
# Use AQI as primary feature
X = df[['index value']]
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
inertia = []

for k in range(1, 10):
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    km.fit(X_scaled)
    inertia.append(km.inertia_)

plt.figure()
plt.plot(range(1, 10), inertia, marker='o')

plt.title("Elbow Method for Optimal Clusters")
plt.xlabel("Number of Clusters")
plt.ylabel("Inertia")

plt.show()
```

```python
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
df['cluster'] = kmeans.fit_predict(X_scaled)

df.head()
```

```python
cluster_summary = df.groupby('cluster')['index value'].agg(['min', 'max', 'mean', 'count'])
print(cluster_summary)
```

```python
centers = scaler.inverse_transform(kmeans.cluster_centers_)

for i, center in enumerate(centers):
    print(f"Cluster {i} center AQI:", round(center[0], 2))
```

```python
# Sort clusters by mean AQI
cluster_order = df.groupby('cluster')['index value'].mean().sort_values().index

cluster_labels = {
    cluster_order[0]: 'Low Pollution',
    cluster_order[1]: 'Moderate Pollution',
    cluster_order[2]: 'High Pollution'
}

df['pollution_level'] = df['cluster'].map(cluster_labels)

df.head()
```

```python
plt.figure(figsize=(12,5))

for cluster in df['cluster'].unique():
    subset = df[df['cluster'] == cluster]
    plt.scatter(subset['date'], subset['index value'], label=f'Cluster {cluster}')

plt.title("AQI Clusters Over Time")
plt.xlabel("Date")
plt.ylabel("AQI")

plt.legend()
plt.grid()

plt.show()
```

```python
df['pollution_level'].value_counts().plot(kind='bar')

plt.title("Distribution of Pollution Levels")
plt.xlabel("Category")
plt.ylabel("Number of Days")

plt.show()
```

```python
for label in df['pollution_level'].unique():
    subset = df[df['pollution_level'] == label]
    print(f"{label}: {subset['index value'].min()} - {subset['index value'].max()}")
```

```python
plt.figure(figsize=(8,5))

for cluster in df['cluster'].unique():
    subset = df[df['cluster'] == cluster]
    plt.hist(subset['index value'], alpha=0.5, label=f'Cluster {cluster}')

plt.title("AQI Distribution by Cluster")
plt.xlabel("AQI")
plt.ylabel("Frequency")

plt.legend()
plt.show()
```

```python
def aqi_category(aqi):
    if aqi <= 50:
        return "Good"
    elif aqi <= 100:
        return "Moderate"
    else:
        return "Unhealthy"

df['aqi_category'] = df['index value'].apply(aqi_category)

pd.crosstab(df['pollution_level'], df['aqi_category'])
```

### Clustering Insights

- KMeans clustering grouped AQI into three levels:
  Low, Moderate, and High Pollution.

- The Elbow Method confirmed that 3 clusters are optimal.

### Observations

- Most days fall under Moderate Pollution.
- High pollution occurs less frequently but shows spikes.
- Low pollution days indicate cleaner air periods.

### Conclusion

Clustering helps identify pollution severity patterns and can assist in
monitoring environmental risk levels and planning mitigation strategies.
