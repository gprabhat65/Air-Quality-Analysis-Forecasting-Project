```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/Delhi_AQIBulletins.csv')

# Standardize column names (VERY IMPORTANT)
df.columns = df.columns.str.strip().str.lower()

df.head()
```

```python
print("Columns in dataset:\n", df.columns)
```

```python
df.info()
df.describe()
```

```python
# Convert date column
df['date'] = pd.to_datetime(df['date'], errors='coerce')

# Sort values
df = df.sort_values('date')

# Reset index
df = df.reset_index(drop=True)

df.head()
```

```python
print("Missing values before cleaning:\n")
print(df.isnull().sum())
```

```python
# Fill AQI missing values
df['index value'] = df['index value'].ffill()

# Drop rows where date is missing
df = df.dropna(subset=['date'])

print("\nMissing values after cleaning:\n")
print(df.isnull().sum())
```

```python
plt.figure(figsize=(12,5))
plt.plot(df['date'], df['index value'])

plt.title("AQI Trend Over Time - Coimbatore")
plt.xlabel("Date")
plt.ylabel("Index Value")

plt.grid()
plt.show()
```

```python
# Using statistical method (more professional)
q1 = df['index value'].quantile(0.25)
q3 = df['index value'].quantile(0.75)
iqr = q3 - q1

upper_limit = q3 + 1.5 * iqr

outliers = df[df['index value'] > upper_limit]

print("Number of outliers:", len(outliers))
outliers.head()
```

```python
df['prominent pollutant'].value_counts().plot(kind='bar')

plt.title("Distribution of Prominent Pollutants")
plt.xlabel("Pollutant")
plt.ylabel("Count")

plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['index value'], bins=30)

plt.title("AQI Value Distribution")
plt.xlabel("Index Value")
plt.ylabel("Frequency")

plt.show()
```

### Key Observations

1. AQI values mostly range between 40–100, indicating moderate pollution levels.
2. Few high AQI spikes (outliers) are observed.
3. Pollution levels show variation over time.
4. The dominant pollutant is PM2.5 (based on dataset).
5. Data cleaning successfully handled missing values.
